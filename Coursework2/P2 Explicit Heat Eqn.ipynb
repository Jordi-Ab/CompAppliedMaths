```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as pt
%matplotlib inline
```

Here we use the Explicit Method to solve the parabolic PDE (heat equation) for $u(t,x) \in \mathbb{R}$ subject to Dirichlet BC's and initial condition:


$U_t - aU_{xx}= f(x, t) \space \space \text{for}\space t \in (0,T], \space x \in (0,1)\\$

$U(t,0)=g_0,\\$
$U(t,1)=g_1,\\$
$U(0,x)=U_0(x)\\$


with $a>0$ (heat-conduction coefficient), $T>0$ and $U_0: (0,1) \rightarrow \mathbb{R}$

```python
def true_sol(t, x):
    #return np.exp(-t)*np.sin(np.pi*x) + np.exp(-4*t)*np.sin(2*np.pi*x) # Epperson Example
    return np.exp(-4*t)*np.sin(2*np.pi*x) + x # Function on CW
    #return np.exp(-(np.pi**2)*t)*np.sin(np.pi*x) # Epperson excercise
    
def rhs_f(t, xvec):
    return np.zeros(xvec.size)
```

```python
n_x = 16
n_t = 64
T = 1
a = 1/(np.pi**2)
#a = 1
g0 = 0
g1 = 1

#def u_0(x): return np.sin(np.pi*x) + np.sin(2*np.pi*x) # Epperson Example
def u_0(x): return np.sin(2*np.pi*x) + x # Function on CW
#def u_0(x): return np.sin(np.pi*x) # Epperson excercise
```

```python
xs = np.linspace(0,1, n_x+1)
h = xs[1] - xs[0]
ts = np.linspace(0,T, n_t+1)
dt = ts[1] - ts[0]
```

```python
fig = pt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(xs, u_0(xs), 'ro', label = 'computed')
ax.plot(xs, true_sol(0, xs), 'b-', label = 'true')
```

```python
# Courant-Friedrichs-Loewy coefficient.
cfl = a * dt / h / h
if (cfl >= 0.5 ): print("Warning, method is unstable for these values of dt and h.")
```

```python
# Inner tridiagonal matrix
first_row = np.zeros(n_x-1)
first_row[0] = 2
first_row[1] = -1
K = la.toeplitz([first_row])
#print(K)
```

```python
I = np.eye(n_x-1)
```

```python
uh = u_0(xs[1:-1])
for t in ts:
    
    # Right Hand Side
    rhs = rhs_f(t,xs[1:-1])
    rhs[0] = rhs[0] + (a/(h**2))*g0
    rhs[-1] = rhs[-1] + (a/(h**2))*g1
    
    sol = np.zeros(n_x+1)
    sol[0] = g0
    sol[1:-1] = uh
    sol[-1] = g1
    
    u_true = true_sol(t, xs)
    
    if(t == 0 or t == 0.5 or t==1):
        fig = pt.figure()
        ax = fig.add_subplot(1,1,1)
        ax.plot(xs, sol, 'ro', label = 'computed')
        ax.plot(xs, u_true, 'b-', label = 'true')
        ax.set_title(r"Heat Equation @ time: "+str(t))
        ax.legend(loc='upper left')
        ax.set_xlabel('x space')
        ax.set_ylabel('u(x,'+str(t)+') state')
        
    uh = (I - cfl*K).dot(uh) + rhs*dt
    
```

```python
sol = np.zeros(n_x+1)
sol[0] = g0
sol[1:-1] = u
sol[-1] = g1
```

```python
pt.plot(xs, true_sol(1,xs))
pt.plot(xs, sol)
```

```python
def mult_tridiagonal(diag, offd, v):
    #Multiply first_row
    n = diag.size
    result = np.zeros(n)
    result[0] = diag[0]*v[0] + offd[0]*v[1]
    for i in  range(offd.size-1):
        result[i+1] = offd[i]*v[i]+diag[i+1]*v[i+1]+offd[i]*v[i+2]
    result[n-1] = offd[n-2]*v[n-2] + diag[n-1]*v[n-1]
    return result
```
